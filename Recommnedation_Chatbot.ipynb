### Read the data

```python
from langchain.document_loaders import DirectoryLoader
loader = DirectoryLoader('/content/',glob='*.txt')
loader = loader.load()
loader
```

### Chunk and vector store

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_spliter = RecursiveCharacterTextSplitter(chunk_size=1000)
text_spliter = text_spliter.split_documents(loader)
text_spliter
```

```python
from langchain_community.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
db = FAISS.from_documents(documents=text_spliter, embedding= HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2"))
```

```python
retriever = db.as_retriever(search_type='similarity',search_kwargs={"k":10})
```

### Chat Prompt and Model

```python
from langchain_groq import ChatGroq
GROQ_API_KEY = "GROQ_API_KEY"
model =  ChatGroq(temperature=0, model_name="llama-3.1-8b-instant", groq_api_key=GROQ_API_KEY)
```

```python
from langchain_core.prompts import ChatPromptTemplate

system_message = """
Provide detailed and personalized course recommendations based on the user's background, interests, and prior learning. If relevant courses are not available in the provided context, just say, "No relevant courses found in the context." Do not suggest incorrect or unrelated courses.\n\n
Context:\n {context}?\n
"""

chat_prompt = ChatPromptTemplate(
    [('system', system_message),
     ("human",'{input}')]
)
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain

question_answer_chain = create_stuff_documents_chain(model, chat_prompt)
```

```python
from langchain.chains import create_retrieval_chain

retrieval = create_retrieval_chain(retriever, question_answer_chain)
```

```python
input = "Who is Ye Bhone Lin"

message1 = retrieval.invoke({"input": input})
```

```python
message1['answer']
```

### Load the model

```python
from langchain_community.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = FAISS.load_local("/content/faiss_index", embedding_model,allow_dangerous_deserialization=True)
```

```python
import pickle

with open("/content/rag_components.pkl", "rb") as f:
    loaded_data = pickle.load(f)

retriever2 = loaded_data["retriever"]
chat_prompt2 = loaded_data["chat_prompt"]

```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain.chains import create_retrieval_chain

question_answer_chain = create_stuff_documents_chain(model, chat_prompt2)
retrieval = create_retrieval_chain(retriever2, question_answer_chain)
```

```python
input = "I want to be become the AI engineer but i havent started my career yet. How can i start"
message1 = retrieval.invoke({"input": input})
```
